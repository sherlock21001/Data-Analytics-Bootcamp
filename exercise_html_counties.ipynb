```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland#List_of_counties"
```

```python
# connect to the website
# website returns a response

response = requests.get(URL)
```

```python
# check the status code
response.status_code
```

```python
# display the first 300 characters in the string
print(response.text[:300])
```

```python
# parse the string as HTML
raw_HTML = BeautifulSoup(response.text, 'html.parser')
```

```python
# search for the <table> tag on the page
#raw_HTML.find('table')
#ignore this line for this exercise
```

```python
# can also search the same table using its class type
raw_HTML.find('table', class_= 'wikitable sortable')
```

```python
# save table to variable
table = raw_HTML.find('table', class_='wikitable sortable')
```

```python
# find_all creates a list of all <tr> tags in table

tabledata = table.find_all('tr')[1:]
```

```python
tabledata
```

```python
type(tabledata)
```

```python
# Create a dictionary with a list of information for us.
state_info = {'County Name':[],
              'FIPS Code':[],
              'County Seat':[],
              'Established (year)':[],
              'Origin':[],
              'Etymology':[],
              "Flag":[],
              "Seal":[],
              'Population':[],
              'Area':[],}
```

```python
# Make sure it's empty
state_info.values()
```

```python
# convert the table headers in each row with th to td
# https://beautiful-soup-4.readthedocs.io/en/latest/#kinds-of-objects
for row in tabledata:
    row_header = row.find('th')
    row_header.name = "td"
```

```python
tabledata
```

```python
# loop through each <td> tag
for row in tabledata:
    # get a list of <td> tags for that row
    row_data = row.find_all('td')
    # get a list of the values (empty lists) from the dictionary
    # loop through the index number and each item (empty list)
    for index, item in enumerate(state_info.values()):
        
        # access the data from the <td> tag with the same index position as the empty list
        data = row_data[index].text.strip()
        #check for the text that couldn't render correctly
        if "\xa0" in data:
            data = data.replace("\xa0"," ")
        # check if the data is a newline or empty string
        # if so, store a null value
        if (data == "\n") or (data == ''):
            item.append(np.nan)
        
        # otherwise, store the actual data from that cell
        else:
            item.append(data)
```

```python
state_info
```

```python
del state_info["Seal"]
del state_info["Flag"]
```

```python
df = pd.DataFrame(data = state_info)
df
```

```python
df.to_csv('countiesMD.csv',index=False)
```
